```python
username = 'root'
password = 'toor'
host = 'localhost'
database = 'sakila'

engine = create_engine(f'mysql+mysqlconnector://{username}:{password}@{host}/{database}')
```

```python
query = """
SELECT 
first_name,
last_name,
email,
address,
city
FROM sakila.customer a
JOIN sakila.address b ON a.address_id = b.address_id
JOIN sakila.city c ON b.city_id = c.city_id
WHERE c.city_id = 321
"""
pd.read_sql(query, engine)
```

```python
query = """
select
title,
description,
release_year,
rating,
special_features
from sakila.film a

JOIN sakila.film_category b on a.film_id = b.film_id
JOIN sakila.category c on b.category_id = c.category_id
where c.category_id = 5
"""
pd.read_sql(query, engine)
```

```python
query = """
SELECT 
last_name,
title,
release_year
FROM sakila.film a
JOIN sakila.film_category b ON a.film_id = b.film_id
JOIN sakila.category c ON b.category_id = c.category_id
JOIN sakila.film_actor d ON a.film_id = d.film_id
JOIN sakila.actor e ON d.actor_id = e.actor_id
WHERE e.actor_id = 5
LIMIT 0, 1000;
"""
pd.read_sql(query, engine)
```

```python
query = """
SELECT 
first_name,
last_name
FROM sakila.film a
JOIN sakila.film_category b ON a.film_id = b.film_id
JOIN sakila.category c ON b.category_id = c.category_id
JOIN sakila.film_actor d ON a.film_id = d.film_id
JOIN sakila.actor e ON d.actor_id = e.actor_id
WHERE title = 'Bingo Talented'
"""
pd.read_sql(query, engine)
```

```python
query = """
SELECT customer_id, amount
FROM sakila.payment
WHERE amount > 2 * (SELECT AVG(amount) FROM sakila.payment);
"""
pd.read_sql(query, engine)
```
